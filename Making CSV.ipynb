```python
import pandas as pd
import os
import json
```

```python
df_l = []
```

```python
for file_name in os.listdir('./processed_data/matches/'):
    with open(os.path.join('./processed_data/matches/', file_name)) as f:
        s = json.load(f)
        df_l.append(s)
```

```python
df = pd.DataFrame.from_records(df_l)
```

```python
df
```

```python
df.to_csv('matches.csv', index=False)
```
